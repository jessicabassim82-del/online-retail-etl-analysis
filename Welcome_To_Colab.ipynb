```
import pandas as pd
import sqlite3
from google.colab import files
```

```
df = pd.read_csv("/content/online_retail_II.csv", on_bad_lines='skip', encoding='ISO-8859-1')
df.head()
```

```
df.dropna(inplace=True)


if 'InvoiceDate' in df.columns:
    df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')


if 'Quantity' in df.columns and 'Price' in df.columns:
    df['Total'] = df['Quantity'] * df['Price']


conn = sqlite3.connect('online_retail.db')
df.to_sql('sales', conn, if_exists='replace', index=False)




query_top_products = """
SELECT Description, SUM(Total) as TotalSales
FROM sales
GROUP BY Description
ORDER BY TotalSales DESC
LIMIT 10;
"""


query_monthly_revenue = """
SELECT strftime('%Y-%m', InvoiceDate) as Month, SUM(Total) as Revenue
FROM sales
GROUP BY Month
ORDER BY Month;
"""


top_products = pd.read_sql(query_top_products, conn)
monthly_revenue = pd.read_sql(query_monthly_revenue, conn)

print("Top 10 Products by Sales:")
display(top_products)

print("Monthly Revenue Trends:")
display(monthly_revenue)
```
